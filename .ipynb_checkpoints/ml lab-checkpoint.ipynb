```python
def train(con, tar):
    for i, val in enumerate(tar):
        if val == "yes":
            specific_h = con[i].copy()
            break

    for i, val in enumerate(con):
        if tar[i] == "yes":
            for x in range(len(specific_h)):
                if val[x] != specific_h[x]:
                    specific_h[x] = '?'
                else:
                    pass
    return specific_h
```
